```python
import numpy as np #import numpy
import matplotlib.pyplot as plt #import matplotlib

f = lambda x ,y : np.sin(x)*np.cos(y) #we use the lambda function to define an anonymous function for a short period of time
x = np.linspace(-2,2,300) #domain for x from the left to the right in 300 steps
y = np.linspace(-2,2,300) #domain for y from left to right in 300 steps
X, Y = np.meshgrid(x,y) #this forms a matrix of all possible combinations of x and y
F = f(X,Y) #f acts on all the ordered pairs

plt.contour(X,Y,F,100) #do a contour plot with 100 levels
plt.show()
```

```python
#this allows me to plot with any view (but static)

import numpy as np #import numpy
import matplotlib.pyplot as plt #import matplotlib
from mpl_toolkits.mplot3d import Axes3D #mpl = matplotlib. We import Axes3D
from matplotlib import cm #colourmap normalization

f = lambda x ,y : np.sin(x)*np.cos(y) #we use the lambda function to define an anonymous function for a short period of time
x = np.linspace(-2,2,300) #domain for x from the left to the right in 300 steps
y = np.linspace(-2,2,300) #domain for y from left to right in 300 steps
X, Y = np.meshgrid(x,y) #this forms a matrix of all possible combinations of x and y
F = f(X,Y) #f acts on all the ordered pairs

#we are going to wrap the entire plot process inside a function
def plotter(E,A): #E for elevation angle and A for azimuthal angle
    fig = plt.figure(figsize = [18,12]) #set the figure size
    ax = fig.gca(projection = '3d') #gca = get current axes. set up the axes. This comes from Axes3D
    ax.plot_surface(X,Y,F, cmap = cm.coolwarm)
    ax.view_init(elev =E, azim=A)
    plt.show()
    
plotter(45,45) #azimuthal angle 45, elevation angle 45
```

```python
#let us now plot curves in space

import numpy as np
import matplotlib.pyplot as plt


plt.rcParams['legend.fontsize'] = 10




# Prepare arrays x, y, z
r = 2
theta = np.linspace(-4 * np.pi, 4 * np.pi, 100) #curve parameter from -4pi to 4pi in 100 steps
z = 5*theta
x = r * np.sin(theta)
y = r * np.cos(theta)

#we are going to wrap the entire plot process inside a function
def plotter(E,A): #E for elevation angle and A for azimuthal angle
    fig = plt.figure(figsize = [18,12]) #set the figure size
    ax = fig.gca(projection = '3d') #gca = get current axes. set up the axes. This comes from Axes3D
    ax.plot(x, y, z, label='Helix of radius 2')
    ax.view_init(elev =E, azim=A) #set the variables for elevation and azimuthal angle
    ax.legend()
    plt.show()
    

from ipywidgets import interactive
iplot = interactive(plotter,E = (-90,90,5),A = (-90,90,5)) # elevation and azimuthal -90 to 90 by 5 degree steps

iplot



```

```python
#now we want to plot multiple surfaces on the same axes
import numpy as np #import numpy
import matplotlib.pyplot as plt #import matplotlib
from mpl_toolkits.mplot3d import Axes3D #mpl = matplotlib. We import Axes3D
from matplotlib import cm #colourmap normalization

f =  lambda x,y: np.exp(x)*np.cos(y) #we use the lambda function to define an anonymous function for a short period of time
g = lambda x,y: (x+y)/2
x = np.linspace(-2,2,300) #domain for x from the left to the right in 300 steps
y = np.linspace(-2,2,300) #domain for y from left to right in 300 steps
X, Y = np.meshgrid(x,y) #this forms a matrix of all possible combinations of x and y
F = f(X,Y) #f acts on all the ordered pairs
G = g(X,Y)

#we are going to wrap the entire plot process inside a function
def plotter(E,A): #E for elevation angle and A for azimuthal angle
    fig = plt.figure(figsize = [18,12]) #set the figure size
    ax = fig.gca(projection = '3d') #gca = get current axes. set up the axes. This comes from Axes3D
    ax.plot_surface(X,Y,F, cmap = cm.coolwarm)
    ax.plot_surface(X,Y,G, cmap = cm.inferno_r)
    ax.view_init(elev =E, azim=A)
    plt.show()
    
plotter(45,45)
from ipywidgets import interactive
iplot = interactive(plotter,E = (-180,180,10),A = (-180,180,10)) # elevation and azimuthal -180 to 180 by 10 degree steps

iplot
```

```python
#now we want to plot multiple surfaces on the same axes
import numpy as np #import numpy
import matplotlib.pyplot as plt #import matplotlib
from mpl_toolkits.mplot3d import Axes3D #mpl = matplotlib. We import Axes3D
from matplotlib import cm #colourmap normalization

f =  lambda x,y: 2*x**2 + y**2 #we use the lambda function to define an anonymous function for a short period of time
g = lambda x,y: 4*x+2*y+3
x = np.linspace(-2,2,300) #domain for x from the left to the right in 300 steps
y = np.linspace(-2,2,300) #domain for y from left to right in 300 steps
X, Y = np.meshgrid(x,y) #this forms a matrix of all possible combinations of x and y
F = f(X,Y) #f acts on all the ordered pairs
G = g(X,Y)

#we are going to wrap the entire plot process inside a function
def plotter(E,A): #E for elevation angle and A for azimuthal angle
    fig = plt.figure(figsize = [18,12]) #set the figure size
    ax = fig.gca(projection = '3d') #gca = get current axes. set up the axes. This comes from Axes3D
    ax.plot_surface(X,Y,F, cmap = cm.coolwarm)
    ax.plot_surface(X,Y,G, cmap = cm.inferno_r)
    ax.view_init(elev =E, azim=A)
    plt.show()
    

from ipywidgets import interactive
iplot = interactive(plotter,E = (-180,180,5),A = (-180,180,5)) # elevation and azimuthal -180 to 180 by 10 degree steps

iplot
```
