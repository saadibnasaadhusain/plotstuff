```python
#three dimensions
from IPython.display import display, Math, Latex
from mpl_toolkits.mplot3d import axes3d #mpl = matplotlib. We import Axes3D
import matplotlib.pyplot as plt #import matplotlib
import numpy as np #import numpy

print("We want to plot the vector field:")
display(Math('\sin(\pi x) \cos(\pi y) \cos(\pi z) \hat{i} + \sin(\pi x) \sin(\pi y) \cos(\pi z) \hat{j} +\cos(\pi x) \cos(\pi y) \sin(\pi z) \hat{k}'))
x, y, z = np.meshgrid(np.arange(1, 2, 0.1),
                      np.arange(1, 2, 0.1),
                      np.arange(1, 2, 0.1)) #define a meshgrid with x, y and z

u = np.sin(np.pi * x) * np.cos(np.pi * y) * np.cos(np.pi * z)
v = -np.cos(np.pi * x) * np.sin(np.pi * y) * np.cos(np.pi * z)
w = (np.cos(np.pi * x) * np.cos(np.pi * y) *
     np.sin(np.pi * z)) #define your vector fields



def plotter(E,A): #E for elevation and A for azimuthal angle
    fig = plt.figure(figsize = [25,15])
    ax = fig.gca(projection='3d')
    ax.quiver(x, y, z, u, v, w, length=0.1, color = 'green')
    ax.view_init(elev =E, azim=A)
    ax.set_title(' This is the vector field $\sin(\pi x) \cos(\pi y) \cos(\pi z) \hat{i} + \sin(\pi x) \sin(\pi y) \cos(\pi z) \hat{j} +\cos(\pi x) \cos(\pi y) \sin(\pi z) \hat{k}$', {'fontsize': 'large',
 'fontweight' : 'heavy',
 'verticalalignment': 'baseline',
 'horizontalalignment': 'center', 'fontsize': 20})
    plt.show()
    
from ipywidgets import interactive
iplot = interactive(plotter,E = (-90,90,15),A = (-90,90,15)) # elevation and azimuthal -90 to 90 by 5 degree steps

iplot
```
