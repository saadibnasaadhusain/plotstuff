```python
#three dimensions
from mpl_toolkits.mplot3d import axes3d #mpl = matplotlib. We import Axes3D
import matplotlib.pyplot as plt #import matplotlib
import numpy as np #import numpy
%matplotlib inline

x, y, z = np.meshgrid(np.arange(1, 2, 0.1),
                      np.arange(1, 2, 0.1),
                      np.arange(1, 2, 0.1)) #define a meshgrid with x, y and z

u = np.sin(np.pi * x) * np.cos(np.pi * y) * np.cos(np.pi * z)
v = -np.cos(np.pi * x) * np.sin(np.pi * y) * np.cos(np.pi * z)
w = (np.sqrt(2.0 / 3.0) * np.cos(np.pi * x) * np.cos(np.pi * y) *
     np.sin(np.pi * z)) #define your vector fields



def plotter(E,A): #E for elevation and A for azimuthal angle
    fig = plt.figure(figsize = [18,12])
    ax = fig.gca(projection='3d')
    ax.quiver(x, y, z, u, v, w, length=0.1, color = 'green')
    ax.view_init(elev =E, azim=A)
    plt.show()
    
from ipywidgets import interactive
iplot = interactive(plotter,E = (-90,90,5),A = (-90,90,5)) # elevation and azimuthal -90 to 90 by 5 degree steps

iplot
```

```python
#now we make an interactive plot
import numpy as np #import numpy
import matplotlib.pyplot as plt #import matplotlib
from mpl_toolkits.mplot3d import Axes3D #mpl = matplotlib. We import Axes3D
from matplotlib import cm #colourmap normalization

f = lambda x ,y : np.sin(x)*np.cos(y) #we use the lambda function to define an anonymous function for a short period of time
x = np.linspace(-2,2,300) #domain for x from the left to the right in 300 steps
y = np.linspace(-2,2,300) #domain for y from left to right in 300 steps
X, Y = np.meshgrid(x,y) #this forms a matrix of all possible combinations of x and y
F = f(X,Y) #f acts on all the ordered pairs

#we are going to wrap the entire plot process inside a function
def plotter(E,A): #E for elevation angle and A for azimuthal angle
    fig = plt.figure(figsize = [18,12]) #set the figure size
    ax = fig.gca(projection = '3d') #gca = get current axes. set up the axes. This comes from Axes3D
    ax.plot_surface(X,Y,F, cmap = cm.coolwarm)
    ax.view_init(elev =E, azim=A)
    plt.show()
    

from ipywidgets import interactive
iplot = interactive(plotter,E = (-90,90,5),A = (-90,90,5)) # elevation and azimuthal -90 to 90 by 5 degree steps

iplot
```
